```python
import matplotlib.pyplot as plt
import json
import seaborn as sns`
```

```python
patients = [
    "Abdomen",
    "Head_Neck",
    "Lung",
    "Prostate"
]
```

```python
plans = {
    "Abdomen":[
            "JSu-IM103",
            "JSu-IM105"
    ],
    "Head_Neck": [
        "JSu-IM101",
        "JSu-IM102",
        "JSu-IM103",
        "JSu-VA104",
        "JSu-VA105"
    ],
    "Lung": [
       "JSu-IM102",
       "JSu-IM105",
       "JSu-IM107"
    ],
    "Prostate": [
        "C1-IM107",
        "C1-IM108",
        "C1-VA109"
    ]
}
```

```python
tissues = {
    "Abdomen": [
        "Bladder",
        "Body",
        "Bowel",
        "Femur_L",
        "Femur_R",
        "PTV_45",
        "PTV_50.4",
        "Rectum"
    ],
    "Head_Neck": [
        "Body",
        "BrainStem",
        "Larynx",
        "Lips",
        "Mandible",
        "Parotid Right",
        "PTV56",
        "PTV63",
        "PTV70",
        "Spinal Cord"
    ],
    "Lung": [
        "Body",
        "Heart",
        "Lung left",
        "Lung right",
        "Lung-GTV",
        "PTV_63",
        "Spinal cord" 
    ],
    "Prostate": [
        "Bladder",
        "Body",
        "prostate_bed",
        "PTV_56",
        "PTV_68",
        "Rectum"     
    ],
}
```

```python
sns.set_palette("Dark2")
sns.set_context('poster')
```

```python
patient = 'Head_Neck'
tissue = 'Spinal Cord'
plt.figure(figsize=(15, 8))
for plan in plans[patient]:
    with open('{}/plans/{}/curves/{}'.format(patient, plan, tissue), 'r') as file:
        curve_points = json.load(file)['CurvePoints']
    x = [int(point['Dose']) for point in curve_points]
    y = [int(point['Volume']) for point in curve_points]
    plt.plot(x, y, label=plan)
plt.vlines(45, 0, 100, color='r', linestyle='--')
plt.vlines(50, 0, 100, color='r')
name = 'Head_Neck, Spinal Cord'
plt.fill_between([0, 45], [100, 100], color='#87c4ff', alpha=0.2, 
                 label='acceptable area')
plt.fill_between([0, 50], [100, 100], color='#87c4ff', alpha=0.2)
plt.legend(loc='lower left')
plt.title(name)
plt.xlabel('Dose, Gy')
plt.ylabel('Volume, %')
plt.grid()
plt.savefig(('Comapison/{}.png').format('Head_Neck_Spinal_Cord'), dpi=100, bbox_inches='tight')
```

```python
for patient in patients:
    for plan in plans[patient]:
        plt.figure(figsize=(15, 8))
        for tissue in tissues[patient]:
            with open('{}/plans/{}/curves/{}'.format(patient, plan, tissue), 'r') as file:
                curve_points = json.load(file)['CurvePoints']
            x = [int(point['Dose']) for point in curve_points]
            y = [int(point['Volume']) for point in curve_points]
            plt.plot(x, y, label=tissue)
        name = '{}, {}'.format(patient, plan)
        plt.legend()
        plt.title(name)
        plt.xlabel('Dose, Gy')
        plt.ylabel('Volume, %')
        plt.grid()
        plt.savefig(('{}/plans/{}.png').format(patient, plan), dpi=100, bbox_inches='tight')
```
